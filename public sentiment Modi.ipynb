```python
import re
def clean_tweet(tweet):
    tweet = re.sub('http\S+\s*', '', tweet)  # remove URLs
    tweet = re.sub('RT|cc', '', tweet)  # remove RT and cc
    tweet = re.sub('#\S+', '', tweet)  # remove hashtags
    tweet = re.sub('@\S+', '', tweet)  # remove mentions
    tweet = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), '', tweet)  # remove punctuations
    tweet = re.sub('\s+', ' ', tweet)  # remove extra whitespace
    tokens = tweet.split(" ")
    tweet = [ x for x in tokens if len(x) < 25 ]
    tweet = " ".join(tweet)
    return tweet
result['text'] = result["text"].apply(clean_tweet)
```

```python
result
```

```python
def remove_emoji(txt):
    emoji_pattern = re.compile("["
                u"\U0001F600-\U0001F64F"  # emoticons
                u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                u"\U0001F680-\U0001F6FF"  # transport & map symbols
                u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                u"\U00002702-\U000027B0"
                u"\U000024C2-\U0001F251"
                u"\U0001f926-\U0001f937"
                u'\U00010000-\U0010ffff'
                u"\u200d"
                u"\u2640-\u2642"
                u"\u2600-\u2B55"
                u"\u23cf"
                u"\u23e9"
                u"\u231a"
                u"\u3030"
                u"\ufe0f"
                "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', txt)
result['text'] = result['text'].apply(lambda x: remove_emoji(x))
```

```python
import re


def cleanTxt(txt):
    
    txt = re.sub(r'@[A-Za-z0-9_:]+', '' , txt)
    
    txt = re.sub(r'#', '' , txt)
    
    txt = re.sub(r'RT[\s]+','',txt)
    
    txt = re.sub(r'https?:\/\/\S+','',txt)
    
    txt = re.sub(r'\d+', '', txt)
    return txt

df['text'] = df["text"].apply(cleanTxt)
```

```python
import string
string.punctuation
```

```python
#Removing punctation
df['text'] = df['text'].str.translate(str.maketrans(" ", " ", string.punctuation))
```

```python
#FOR  SENTIMENT LABELLING WE HAVE TO IMPORT  NLKT 


import nltk
nltk.download('vader_lexicon')
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()
```

```python
def print_sentiment_scores(tweets):
    vadersenti = analyser.polarity_scores(tweets)
    return pd.Series([vadersenti['pos'], vadersenti['neg'], vadersenti['neu'], vadersenti['compound']])
```

```python
df[['vadersenti_pos', 'vadersenti_neg', 'vadersenti_neu', 'vadersenti_compound']]  = df['text'].apply(print_sentiment_scores)
```

```python
df['senti'] = 'neutral'
df.loc[df['vadersenti_compound'] > 0 , 'senti'] = 'positive'
df.loc[df['vadersenti_compound'] < 0,'senti'] = 'negative'

df['vader_possentiment'] = 0
df.loc[df['vadersenti_compound'] > 0 , 'vader_possentiment'] = 1

df['vader_negsentiment'] = 0
df.loc[df['vadersenti_compound'] < 0 , 'vader_negsentiment'] = 1

df['vader_neusentiment'] = 0
df.loc[(df['vadersenti_compound'] <0) & (df['vadersenti_compound'] >=0) , 'vader_neusentiment'] = 1
```
